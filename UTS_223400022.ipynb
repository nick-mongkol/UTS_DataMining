```python
from sklearn.preprocessing import LabelEncoder

dataolahencoding = df.copy()  # Membuat salinan dataset untuk mencegah modifikasi dataset asli

for column in new_var:
    try:
        encoder = LabelEncoder()
        new_column_name = column + '_encod'
        dataolahencoding[new_column_name] = encoder.fit_transform(dataolahencoding[column])
    except TypeError:
        print(f"Column {column} cannot be encoded using LabelEncoder.")
        continue

dataolahencoding = dataolahencoding.drop(['JUMLAH PENDUDUK', 'LUAS WILAYAH (KM)', 'JUMLAH RUMAH', 'TEMBOK', 'JUMLAH SEKOLAH\n(SMP-SMA)', 'JUMLAH RS', 'Usia 15-64 (%)'], axis=1)
dataolahencoding.head()
```
